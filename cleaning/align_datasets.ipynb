```python
from Levenshtein import distance
import json
```

```python
with open('../datasets.json') as json_file:
    datasets = json.load(json_file)
```

```python
d_titles = []
for d in datasets:
    try:
        name_list = []
        name_list.append(d['title'])
        name_list.append(d['alias'])
        d_dict = {'dataset_id':d['dataset_id'],'names':name_list}
        d_titles.append(d_dict)
    except:
        pass
```

```python
incoming_dataset_name = "IRI scanner data (2013) "
# edit_dist = distance("ah", "aho")
dist_list = []
for d in d_titles:
    for n in d['names']:
        dist_dict = {'dataset_id':d,'ldist':distance(incoming_dataset_name,n)}
        dist_list.append(dist_dict)
        
```
